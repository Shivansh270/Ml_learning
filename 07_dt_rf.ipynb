```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import joblib
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
df = pd.DataFrame(data.data,columns= data.feature_names)
```

```python
df['taraget'] = data.target
```

```python
df
```

```python
df
```

```python
x_multiclass = df[['sepal length (cm)',	'sepal width (cm)',	'petal length (cm)',	'petal width (cm)']]
```

```python
x_multiclass
```

```python
y_multiclass = df['taraget']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_multiclass,y_multiclass,test_size=.2, random_state=10)
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
dt_model = DecisionTreeClassifier(criterion='log_loss', max_depth=3)
```

```python
dt_model.fit(x_train,y_train)
```

```python
dt_model.predict(x_test)
```

```python
confusion_matrix(y_test ,dt_model.predict(x_test) )
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,6))
plot_tree(dt_model,feature_names=data.feature_names , class_names= data.target_names,filled=True)
plt.show()
```

```python
print("print the acuraccy of the model \n",classification_report(y_test ,dt_model.predict(x_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100000,criterion='gini')
```

```python
rf_model.fit(x_train,y_train)
```

```python
print("classificaoitn report printing \n",classification_report(y_test , rf_model.predict(x_test)))
```

```python
for i in range(min(5,len(rf_model.estimators_))) : 
    plt.figure(figsize=(12,6))
    plot_tree(rf_model.estimators_[i],feature_names=data.feature_names , class_names= data.target_names,filled=True)
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.datasets import fetch_california_housing
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data.data , columns=data.feature_names)
```

```python
df['target'] = data.target
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
x = df.head(20).drop(columns= ['target'] )
y = df['target'][:20]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , random_state=10 , test_size=.1)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test  = scaler.transform(x_test)
```

```python
x_train
```

```python
rf_model_reg = RandomForestRegressor(n_estimators=1000 , criterion='squared_error' , random_state= 30 )
```

```python
rf_model_reg.fit(x_train , y_train)
```

```python
rf_model_reg.predict(x_test)
```

```python
r2_score(y_test , rf_model_reg.predict(x_test))
```

```python
mean_squared_error(y_test , rf_model_reg.predict(x_test))
```

```python
joblib.dump(rf_model_reg , 'rf_reg_model_1000.pkl')
```

```python
for i in range(min(5,len(rf_model_reg.estimators_))) : 
    plt.figure(figsize=(12,6))
    plot_tree(rf_model_reg.estimators_[i],feature_names=data.feature_names , class_names= data.target_names,filled=True)
    plt.show()
```

```python
from sklearn.svm import SVR
```

```python
df
```

```python
x = df.drop(columns= ['target'])
y = df['target']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size=.2 , random_state= 20 )
```

```python
scaler = StandardScaler()
x_train  = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
svr_resressor_model  = SVR(kernel='rbf' ,C = 100 ,epsilon=.1)
```

```python
svr_resressor_model.fit(x_train ,y_train)
```

```python
r2_score(y_test , svr_resressor_model.predict(x_test))
```

```python
x_single_feature_train = x_train[:,0].reshape(-1,1)
x_single_feature_test = x_test[:,0].reshape(-1,1)
```

```python
x_single_feature_train
```

```python
svr_resressor_model_single_col   = SVR(kernel='linear' ,C = 100 ,epsilon=.1)
```

```python
svr_resressor_model_single_col.fit(x_single_feature_train , y_train)
```

```python
y_pred_single = svr_resressor_model_single_col.predict(x_single_feature_test)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x_single_feature_test[:,0],y_test,color = 'blue',label = 'Actual' , alpha=.5)
plt.scatter(x_single_feature_test[:,0],y_pred_single,color = 'red',label = 'svr_prediction' , alpha=.5)
```

```python
x_train_2d = x_train[: , [0,1]]
x_test_2d = x_test[: , [0,1]]
```

```python
x_train_2d
```

```python
svr_2d = SVR(kernel='linear' , C = 100 , epsilon=.1)
svr_2d.fit(x_train_2d,y_train)
```

```python
y_pred_2d = svr_2d.predict(x_test_2d)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_test_2d[:, 1] , x_test_2d[: , 0] , y_test , color  = 'blue' , label = 'Actual' , alpha = .5)
ax.scatter(x_test_2d[:, 0] , x_test_2d[: , 1] , y_pred_2d , color  = 'red' , label = 'Actual' , alpha = .5)
```
